```python
from torch import tensor
from torch import nn
from torch import sigmoid
import torch.nn.functional as F
import torch.optim as optim
import torch
```

### Using sigmoid

```python
x_data=tensor([[1.0],[2.0],[3.0],[4.0]])
y_data=tensor([[0.0],[0.0],[1.0],[1.0]])
```

```python
class Model(nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.linear=nn.Linear(1,1)
        
    def forward(self, x):
        y_pred=sigmoid(self.linear(x))
        return y_pred
        
# our model
model=Model()
```

```python
criterion=nn.BCELoss(reduction="mean") # softmax를 포함한다.
optimizer=optim.SGD(model.parameters(),lr=0.01)
```

```python
for epoch in range(1000):
    y_pred=model(x_data)
    
    loss=criterion(y_pred,y_data)
    print(f"Epoch{epoch+1} | Loss: {loss.item(): .4f}")
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
hour_var=tensor([[1.0]])
print("predict 1 hour ", 1.0, model(hour_var).data[0][0]>0.5)
hour_var=tensor([[7.0]])
print("predict 7 hour ", 7.0, model(hour_var).data[0][0]>0.5)
```

### Using ReLU

```python
x_data=tensor([[1.0],[2.0],[3.0],[4.0]])
y_data=tensor([0,0,1,1])
```

```python
class Model(nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.linear1=nn.Linear(1,5)
        self.linear2=nn.Linear(5,5)
        self.linear3=nn.Linear(5,2)
        self.act=nn.ReLU()
        
    def forward(self, x):
        x=self.linear1(x)
        x=self.act(x)
        x=self.linear2(x)
        x=self.act(x)
        x=self.linear3(x)
        x=self.act(x)
        return x
model=Model()
criterion=nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters())
```

```python
for epoch in range(1000):
    y_pred=model(x_data)
    loss=criterion(y_pred,y_data)
    print(f"Epoch{epoch+1} | Loss: {loss.item(): .4f}")
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
hour_var=tensor([[1.0]])
print("predict 1 hour ", 1.0, torch.argmax(model(hour_var).data))
hour_var=tensor([[7.0]])
print("predict 7 hour ", 7.0, torch.argmax(model(hour_var).data))
```

### Using LeakyReLU

```python
class Model(nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.linear1=nn.Linear(1,5)
        self.linear2=nn.Linear(5,5)
        self.linear3=nn.Linear(5,2)
        self.act=nn.LeakyReLU()
        
    def forward(self, x):
        x=self.linear1(x)
        x=self.act(x)
        x=self.linear2(x)
        x=self.act(x)
        x=self.linear3(x)
        x=self.act(x)
        return x
model=Model()
criterion=nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters())
```

```python
for epoch in range(1000):
    y_pred=model(x_data)
    loss=criterion(y_pred,y_data)
    print(f"Epoch{epoch+1} | Loss: {loss.item(): .4f}")
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
hour_var=tensor([[1.0]])
print("predict 1 hour ", 1.0, torch.argmax(model(hour_var).data))
hour_var=tensor([[7.0]])
print("predict 7 hour ", 7.0, torch.argmax(model(hour_var).data))
```
